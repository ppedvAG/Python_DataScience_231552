## Plotting

Wir können unsere Daten grafisch darstellen mithilfe von verschiedenen Tools. Das beliebteste Tool ist matplotlib

Pandas hat eine direkte Integration mit matplotlib, dadurch können wir Series/DataFrames direkt in Graphen darstellen

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # Wichtig: Unterpaket pyplot importieren
```

```python
data = pd.read_csv("Data/AmazonStock.csv", delimiter=",", decimal=".", index_col="Date", parse_dates=True)
```

## Objektorierter vs. Globaler Modus

Beim globalen Modus wird mittels plt.plot ein Graph erstellt und dieser wird durch die folgenden Methoden bearbeitet

Beim objektorientierten Modus wird hingegen der Plot in einer Variable gespeichert, und auf diese Variable können die Methoden angewandt werden

```python
a = np.arange(100)
plt.figure(figsize=(12, 6))  # figure: Leere Fläche
plt.title("Linien")
plt.plot(a)
plt.plot(a * 2)
plt.plot(a * 3)
```

```python
x = np.arange(-10, 11)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # subplot(Figure, Plot, Max): Ermöglicht mehrere Plots nebeneinander zu erstellen
plt.plot(x, x ** 2)
plt.plot(x, x ** 3)
plt.legend(["X^2", "X^3"], loc="upper right")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1, 2, 2)
plt.plot(x, -x ** 2)
plt.plot(x, -x ** 3)
plt.plot([-1000, 0, 1000])
plt.legend(["-X^2", "-X^3"])
```

## Arbeiten mit mehreren Plots im OOP Modus

```python
figure, ((p1, p2), (p3, p4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))  # Bei einem Grid, müssen die Variablen als Tupel von Tupeln angegeben werden
```

```python
p1.plot(np.random.randn(100), c="Red")
p2.plot(np.random.randn(100), c="Green")
p3.plot(np.random.randn(100), c="Blue")
p4.plot(np.random.randn(100), c="Yellow")
figure
```

## Bar Chart (Balkendiagramm)

```python
x = np.array([x for x in range(0, 10)])
y = np.array([x for x in range(0, 10)])
```

```python
fig = plt.figure(figsize=(12, 6))
plt.bar(np.arange(len(x)), x, label="X Werte", width=0.4)  # Erforderlich: Array mit Länge x, Die Daten selbst (x)
plt.bar(np.arange(len(y)) + 0.4, y, label="Y Werte", width=0.4)
plt.grid(alpha=0.2, color="Black", axis="y", linestyle="--", linewidth=2)
plt.xlabel("Index")
plt.ylabel("Zahlenwerte")
plt.legend()

None
```

## Histogramm

Zeigt die Menge der Daten (kumuliert) wie bei einem Balkendiagramm an

```python
x = np.array([np.random.randint(0, 1000) for i in range(0, 1000)])
```

```python
plt.figure(figsize=(12, 6))

plt.hist(x, bins=100, edgecolor="#00AA00", color="#BBBBBB")
```

## Pie Chart (Tortendiagramm)

```python
x = np.arange(1, 10)

plt.figure(figsize=(6, 6))

plt.pie(x, radius=2.5, center=(2, 2), frame=True, labels=[f"Zahl: {i}" for i in x], labeldistance=0.7, counterclock=True, rotatelabels=True, startangle=100,
        wedgeprops={"linewidth": 1, "edgecolor": "white"})
```

## Scatterplot (Streudiagramm)

Ein Streudiagramm erlaubt uns, zwei Werte gleichzeitig zu visualisieren die im Verhältnis zueinander stehen.

Wird häufig verwendet um Outlier zu erkennen

```python
x = np.array([np.random.randint(0, 20) + x for x in range(0, 100)])
y = np.array([np.random.randint(0, 5) + x for x in range(0, 100)])
```

```python
col = [x[i] * y[i] for i in range(0, 100)]
area = [i for i in range(0, 100)]

plt.scatter(x, y, c=col, s=area)
plt.colorbar()
```
