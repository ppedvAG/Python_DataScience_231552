## Andere Datenquellen

Es ist auch möglich andere Quellen als CSV zu benutzen (z.B. Json, SQL, ...)

```python
import pandas as pd
import requests as r
```

```python
json = r.get("https://open.er-api.com/v6/latest/USD")
```

```python
json.json()
```

```python
json.json()['rates']
```

```python
jsonDF = pd.DataFrame.from_dict(json.json()['rates'], orient="index")
jsonDF = jsonDF.rename(columns={0: "Exchange Rates"})
jsonDF
```

```python
readJson = pd.read_json("https://open.er-api.com/v6/latest/USD")
```

```python
readJson
```
